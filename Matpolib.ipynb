```python
from skimage import data, color

import matplotlib.pyplot as plt
 
#Charger l'image couleur par defaut de scikit-image(chelsea)

face = data.chelsea()  #charge l'imùage couleur par defaut

#convertir l'image en niveau de gris

face_gray = color.rgb2gray(face)

#Affiche mon image 

plt.imshow(face_gray, cmap=plt.cm.gray)

plt.show()
plt.imshow(face)
```

```python
#Charger l'image couleur par defaut de scikit-image(chelsea)

face = data.chelsea()  #charge l'image couleur par defaut

#convertir l'image en niveau de gris

#face_gray = color.rgb2gray(face)

h= face_gray.shape[0]# hauteur
w= face_gray.shape[1]#largeur
#print(h)
#print(w)

New_face = face_gray[h//4 : -h//4, w//4 : -w//4] #redimensione l'image
New_face[New_face > 30] = 255


#Affiche mon image :

plt.imshow(New_face, cmap=plt.cm.gray)


#plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,2,10)# les option de creation du plot
y = x**2
print(x)
print(y)
```

Pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7, 3))# taille du graphique
plt.plot(x,y, c= 'red', lw= 4 , ls='--', label = 'quadratique')#ajout un graph
plt.plot(x,y**4, lw= 2 , ls='--', label='quartique')#ajout une 2eme courbe sur le meme graph
plt.plot(x,x**9,label= 'autre')#ajout une 2eme courbe sur le meme graph

plt.title('Figure AJC1')
plt.xlabel('Axe des x')
plt.ylabel('Axe des y')
plt.legend()
plt.show()
plt.savefig('AJC1.png')


```

```python
#plt.plot(x,y, c= 'red', lw= 4 , ls='--')#tableau x et y c= colore lw= epasseur ls= style de courbe
#plt.scatter(x,y)
plt.plot(x,y)
plt.show()#affiche notre graphique 
```

```python
plt.subplot(2,2,1)#ajoute 2 lignes et une colonne
plt.plot(x,y,c= 'blue')
plt.subplot(2,2,2)
plt.plot(x,y,c='black')
plt.subplot(2,2,3)
plt.plot(x,y,c='green')
plt.subplot(2,2,4)
plt.plot(x,y,c='brown')



plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.subplot(2, 2, 1)
plt.plot(y, x, c='brown', lw=5, ls='--', label='quadratique')
plt.title('Figure1 AJC')
plt.subplot(2, 2, 2)
plt.plot(y, np.sin(x), label='quartique')
plt.show()
```

```python
#POO subplot avc S
fig , ax = plt.subplots(2,1, sharex= True)
ax[0].plot(x,y)
ax[1].plot(x,y**2)
plt.show()
#ax.shape# donne le nombre dedimmenssion au tableau
```
