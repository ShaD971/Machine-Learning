```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
#Data
X,y = make_blobs(n_samples = 200 , centers=3, cluster_std=0.5, random_state=0)
plt.scatter(X[:,0],X[:,1])
plt.title( 'Nuage de point')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# G√©n√©rer des donn√©es de test
X, y = make_blobs(n_samples=200, centers=3, cluster_std=0.5, random_state=2)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolor='k', s=50)

plt.title('Scatter Plot of Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')


handles, labels = plt.gca().get_legend_handles_labels()
unique_labels = set(y)
legend_labels = [f'Cluster {label}' for label in unique_labels]

plt.legend(handles, legend_labels, title='Clusters')

plt.show()
```

```python
#model = KMeans(n_clusters=3, max_iter=300, n_init = 10, init = K-Means++) #tous les paramettre facultatif
model = KMeans(n_clusters=3)#model de donn√© de depart
model.fit(X)
model.labels_
```

```python
model.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=model.predict(X), cmap='viridis', marker='o', edgecolor='k', s=50)
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c='b') # affiche le centroid
#model.inertia_ #calcukle la plus petit distance entre les point
model.score(X)
```

```python
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1])#position du centroid
```

```python
liste = []
k_al = range(1,30)
for k in k_al :
    model = KMeans(n_clusters=k).fit(X)
    liste.append(model.inertia_)


plt.plot(k_al,liste)
plt.xlabel(' K clusters')
plt.ylabel('Fonction inertia')
plt.show()
```

```python
#isolation Forest
# G√©n√©rer des donn√©es de test
X, y = make_blobs(n_samples=20, centers=1, cluster_std=0.3, random_state=2)
X[-1,: ]= np.array([2.25,5])

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolor='k', s=50)

plt.title('Scatter Plot of Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')


handles, labels = plt.gca().get_legend_handles_labels()
unique_labels = set(y)
legend_labels = [f'Cluster {label}' for label in unique_labels]

plt.legend(handles, legend_labels, title='Clusters')

plt.show()
```
